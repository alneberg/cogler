```python
import pandas as pd
```

```python
df = pd.read_table("data/scg_phylum_proteobacteriaclass_min0.97_max1.03_min500.tab", index_col=0)
```

```python
df_min_3_genera = df[df['Number_genera'] >= 3]
df_min_3_genera
```

```python
cog_columns = [col for col in df_min_3_genera.columns if not col.startswith('Number')]
non_cog_columns = [col for col in df_min_3_genera.columns if col.startswith('Number')]
```

```python
df_cog_sum = df_min_3_genera[cog_columns].sum()
```

```python
non_zero_cogs = list(df_cog_sum[df_cog_sum != 0].index.values)
```

```python
df_min_3_genera[non_cog_columns + non_zero_cogs]
```

```python
df_min_3_genera[non_cog_columns + non_zero_cogs].to_csv("data/scg_phylum_proteobacteriaclass_min0.97_max1.03_min500_filtered_columns.tab", sep='\t')
```

```python
df_min_3_genera[non_cog_columns + non_zero_cogs].T.to_csv("data/scg_phylum_proteobacteriaclass_min0.97_max1.03_min500_filtered_rows.tab", sep='\t')
```
